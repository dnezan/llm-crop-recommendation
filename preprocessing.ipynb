```python
import csv
import re

input_file = 'C:/Users/Dinesh.Sreekanthan/PycharmProjects/streamlit_geoapp/data/89th Gazette Notification.csv'  # Replace with the path to your input CSV file
output_file = 'C:/Users/Dinesh.Sreekanthan/PycharmProjects/streamlit_geoapp/data/data.csv'  # Replace with the desired output CSV file path

# Open input and output files
with open(input_file, 'r', newline='') as file_in, open(output_file, 'w', newline='') as file_out:
    reader = csv.reader(file_in)
    writer = csv.writer(file_out)

    # Iterate through each row in the input CSV
    for row in reader:
        # Replace multiple spaces with a single space in each value of the row
        cleaned_row = [re.sub(r'\s+', ' ', value) for value in row]
        
        # Write the cleaned row to the output CSV
        writer.writerow(cleaned_row)

print("CSV file processed successfully.")
```

```python
import ee
import geemap
import geemap.foliumap as geemap

# Initialize Earth Engine
ee.Initialize()

# Create a geemap Map instance
Map = geemap.Map()

uploaded_file = 'data (26).geojson'
roi = geemap.geojson_to_ee(uploaded_file)

# Load Landsat 9 imagery
start_date = '2023-06-10'
end_date = '2023-09-20'

# Filter Landsat 9 imagery
l9_collection = ee.ImageCollection('LANDSAT/LC09/C02/T2_L2') \
    .filterBounds(roi) \
    .filterDate(start_date, end_date) \
    .sort('CLOUD_COVER') \
    .map(lambda image: ee.Algorithms.Landsat.simpleCloudScore(image).select(['B[0-9]*']))

# Create a median composite from the collection
l9_median = l9_collection.median()

# Add Landsat 9 median composite to the map
Map.addLayer(l9_median, {'bands': ['B4', 'B3', 'B2'], 'min': 0, 'max': 3000}, 'Landsat 9 Imagery')

# Center the map on the ROI
Map.centerObject(roi)

# Set control visibility
Map.setControlVisibility(layerControl=True, fullscreenControl=True, latLngPopup=True)

# Display the map
Map
```

```python
import ee
import geemap

Map = geemap.Map()
collection = ee.ImageCollection('LANDSAT/LC09/C02/T1_L2')
median = collection.median()

def apply_scale_factors(image):
    opticalBands = image.select('SR_B.').multiply(0.0000275).add(-0.2)
    thermalBands = image.select('ST_B.*').multiply(0.00341802).add(149.0)
    return image.addBands(opticalBands, None, True).addBands(thermalBands, None, True)

dataset = apply_scale_factors(median)

vis_natural = {
    'bands': ['SR_B4', 'SR_B3', 'SR_B2'],
    'min': 0.0,
    'max': 0.3,
}

vis_nir = {
    'bands': ['SR_B5', 'SR_B4', 'SR_B3'],
    'min': 0.0,
    'max': 0.3,
}

vis_agri = {
    'bands': ['SR_B6', 'SR_B5', 'SR_B2'],
    'min': 0.0,
    'max': 0.3,
}

Map.addLayer(dataset, vis_natural, 'True color (432)')
Map.addLayer(dataset, vis_nir, 'Color infrared (543)')
Map.addLayer(dataset, vis_agri, 'Agri')
Map
```
